```python
import matplotlib.font_manager as fm

def find_korean_font():
    """
    시스템에 설치된 한글 폰트를 찾는 함수
    """
    # 자주 사용되는 한글 폰트 목록
    korean_fonts = ['Malgun Gothic', '맑은 고딕', 'AppleGothic', 'NanumGothic',
                   '나눔고딕', 'NanumBarunGothic', 'Gulim', '굴림', 'Dotum', '돋움']

    # 시스템에 설치된 폰트 중 한글 폰트 찾기
    available_fonts = [f.name for f in fm.fontManager.ttflist]
    for font in korean_fonts:
        if font in available_fonts:
            return font

    # 설치된 모든 폰트 출력 (디버깅용)
    print("설치된 폰트 목록:")
    for font in available_fonts:
        print(font)

    return None
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_separate_analysis(df):
    korean_font = find_korean_font()
    if korean_font:
        plt.rcParams['font.family'] = korean_font
    else:
        print("Warning: 한글 폰트를 찾을 수 없습니다.")

    plt.rcParams['axes.unicode_minus'] = False

    # 수치형 칼럼 선택
    numeric_columns = [
        'product_name_length',
        'product_description_length',
        'product_photos_qty',
        'product_weight_g',
        'product_length_cm',
        'product_height_cm',
        'product_width_cm'
    ]

    # 1. 수치형 변수 분포 시각화
    plt.figure(figsize=(15, 10))

    for idx, column in enumerate(numeric_columns, 1):
        plt.subplot(4, 2, idx)
        sns.histplot(data=df, x=column, bins=30, kde=True)
        plt.title(f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Count')

    plt.tight_layout()
    plt.show()

    # 2. 범주형 변수 (카테고리) 빈도 시각화
    plt.figure(figsize=(10, 6))

    category_counts = df['product_category_name'].value_counts()

    sns.barplot(x=category_counts.values,
                y=category_counts.index)

    plt.title('Product Category Distribution')
    plt.xlabel('Count')
    plt.ylabel('Category')

    plt.tight_layout()
    plt.show()

    # 3. 기초 통계량 출력
    print("\nNumerical Variables Summary:")
    print(df[numeric_columns].describe().round(2))

    print("\nCategory Distribution:")
    print(df['product_category_name'].value_counts(normalize=True).round(3) * 100, "%")

# 사용 예시:
plot_separate_analysis(df)
```
